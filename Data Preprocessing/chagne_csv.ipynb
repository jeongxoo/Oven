```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import math
import seaborn as sns
```

```python
temp = pd.read_csv('oven_data_0723.csv')
```

```python
temp
```

```python
match_k = ["a", "b", "c", "d", "e", "f"]
cnt = 0
alpha = 0
all_combi_dict = {}
all_combi_win_dict = {}
dict_keys = []
win_keys = []

while (1):
    cnt += 1
    try:
        key = match_k[alpha] + str(cnt)
    except:
        break
        
    newnew = temp[temp["MATCH_KEY"] == key]
    

    at_combination = []
    df_combination = []
    
    if newnew.empty:
        cnt = 0
        alpha += 1
    else:
        at = newnew[newnew["TEAM"] == "at"]
        df = newnew[newnew["TEAM"] == "df"]
#         print(at["COOKIE_IDX"])
#         print(df["COOKIE_IDX"])

        if len(at["COOKIE_IDX"]) > 5 or len(df["COOKIE_IDX"]) > 5:
            continue

        for aat in at["COOKIE_IDX"]:
            at_combination.append(aat)
                        
        for ddf in df["COOKIE_IDX"]:
            df_combination.append(ddf)
            
        try:
            all_combi_dict[tuple(at_combination)] += 1
        except:
            all_combi_dict[tuple(at_combination)] = 0
            dict_keys.append(tuple(at_combination))
            all_combi_dict[tuple(at_combination)] += 1
            
        try:
            all_combi_dict[tuple(df_combination)] += 1
        except:
            all_combi_dict[tuple(df_combination)] = 0
            dict_keys.append(tuple(df_combination))
            all_combi_dict[tuple(df_combination)] += 1
            
            
        
        if sum(at["RESULT"]) == 5:
            try:
                all_combi_win_dict[tuple(at_combination)] += 1
            except:
                all_combi_win_dict[tuple(at_combination)] = 0
                win_keys.append(tuple(at_combination))
                all_combi_win_dict[tuple(at_combination)] += 1
        elif sum(df["RESULT"]) == 5:
            try:
                all_combi_win_dict[tuple(df_combination)] += 1
            except:
                all_combi_win_dict[tuple(df_combination)] = 0
                win_keys.append(tuple(df_combination))
                all_combi_win_dict[tuple(df_combination)] += 1
        
    
    # 멈추기용 하나만 가지고 테스트할거
#     if cnt == 1:
#         break
```

```python
print(len(dict_keys))
print(len(all_combi_dict))
print(len(all_combi_win_dict))
print(len(win_keys))
```

```python
column_name_fst = ["COMBINATION_KEY", "COMBI_INFO", "COMBI_ONE", "COMBI_TWO", "COMBI_THREE", "COMBI_FOUR", "COMBI_FIVE", "WIN_CNT", "LOOSE_CNT", "ALL_CNT"]
combi_one_df = pd.DataFrame(columns = column_name_fst)
to_cnt = 0

for key in dict_keys:
    to_cnt += 1
    com_key = "combi" + str(to_cnt)
    
    try:
        wins = all_combi_win_dict[key]
    except:
        wins = 0
        
    looses = all_combi_dict[key] - wins
    
    list_combi = []
    for i in range(5):
        try:
            list_combi.append(key[i])
        except:
            list_combi.append(-1)
    
    data_row = [com_key, tuple(list_combi), list_combi[0], list_combi[1], list_combi[2], list_combi[3], list_combi[4], wins, looses, all_combi_dict[key]]
    combi_one_df.loc[to_cnt - 1] = data_row
```
