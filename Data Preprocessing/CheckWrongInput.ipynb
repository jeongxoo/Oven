```python
import pandas as pd
import csv
```

```python
origin_path = "oven_data_0708.csv"
csv_to_check = pd.read_csv(origin_path)
key_of_csv = ["TIMESTAMP", "MATCH_KEY", "RESULT", "TEAM", 
              "USER_ID", "USER_TIER", "USER_TROPHY", 
              "USER_ITEM_1", "USER_ITEM_LV_1", "USER_ITEM_RARITY_1", 
              "USER_ITEM_2", "USER_ITEM_LV_2", "USER_ITEM_RARITY_2", 
              "USER_ITEM_3", "USER_ITEM_LV_3", "USER_ITEM_RARTY_3", 
              "TOTAL_POWER", "MATCH_COOKIE_NUM", "HIDDEN_COOKIE", 
              "COOKIE_NAME", "COOKIE_IDX", "COOKIE_POS", "COOKIE_STAR", "COOKIE_RARITY",
              "COOKIE_DEFAULT_POS", "COOKIE_TYPE", "COOKIE_POWER", "COOKIE_LV", "COOKIE_SK_LV", 
              "GIVE_DMG", "RCV_DMG", "GIVE_HEAL", "RCV_HEAL"]
which_is_integer = [2, 6, 8, 11, 14, 16, 17, 18, 20, 21, 22, 26, 27, 28, 29, 30, 31, 32]
```

```python
csv_to_check
```

```python
def solStrToInt(sen):
    new = ""
    for s in sen:
        try:
            int(s)
            new = new + s
        except:
            print("%s에서 문제가 되는 \"%s\" 제거 후" % (sen, s), end = "")
    
    return new
```

```python
def changeToNewCSV(csv_to_check, key_of_csv, which_is_integer):
    for i in range(len(csv_to_check)):
    #     print("*")
        for j in range(len(key_of_csv)):
            if pd.isna(csv_to_check[key_of_csv[j]][i]):       
                print("공백 오류 발생")
                print("%d번째 행, %d번째 열" % (i + 2, j + 1))
                print("*" * 20)
                print("")
            else:
                 if (j in which_is_integer):
                        try:
                            int(csv_to_check[key_of_csv[j]][i])
                        except:
                            print("형변환 오류 발생")
                            print(type(csv_to_check[key_of_csv[j]][i]))
                            print("%d번째 행, %d번째 열" % (i + 2, j + 1))
                            a = solStrToInt(csv_to_check[key_of_csv[j]][i])
                            csv_to_check.loc[(i, key_of_csv[j])] = a
                            # https://pandas.pydata.org/pandas-docs/stable/user_guide/indexing.html#returning-a-view-versus-a-copy
                            print(" %s로 변경" %csv_to_check[key_of_csv[j]][i])
                            print("*" * 20)
                            print("")
    
    return csv_to_check
```

```python
new = changeToNewCSV(csv_to_check, key_of_csv, which_is_integer)
```
